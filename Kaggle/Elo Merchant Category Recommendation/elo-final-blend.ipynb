```python
import numpy as np # NUMPY
import pandas as p # PANDAS

# DATA VIZUALIZATION LIBRARIES
from matplotlib import pyplot as plt
import seaborn as sns

# METRICS TO MEASURE RMSE
from math import sqrt
from sklearn import metrics
```

```python
df_base0 = p.read_csv('../input/elo-final-blend-dataset/combining_submission.csv',names=["card_id","target0"], skiprows=[0],header=None)
df_base1 = p.read_csv('../input/elo-final-blend-dataset/output-ashish-3.691.csv',names=["card_id","target1"], skiprows=[0],header=None)
df_base2 = p.read_csv('../input/elo-final-blend-dataset/subm_3.648358_LGBM_cv11_2019-02-19-04-43.csv',names=["card_id","target2"], skiprows=[0],header=None)
df_base3 = p.read_csv('../input/elo-final-blend-dataset/subm_3.651336_XGB_cv11_2019-02-19-06-09.csv',names=["card_id","target3"], skiprows=[0],header=None)
df_base4 = p.read_csv('../input/elo-final-blend-dataset/subm_3.651336_XGB_cv11_2019-02-19-06-10.csv',names=["card_id","target4"], skiprows=[0],header=None)
```

```python
df_base_final = p.read_csv('../input/elo-final-blend-dataset/combining_submission.csv',names=["card_id","target"], skiprows=[0],header=None)
```

```python
df_base_final['target'] = .3*df_base0['target0'] + .6*df_base2['target2']  + .0.1*df_base1['target1'] # + .2*df_base2['target2'] 
```

```python
df_base_final.to_csv('df_base_final.csv',index=False)
```

```python
df_base = p.merge(df_base1,df_base0,how='inner',on='card_id')
df_base = p.merge(df_base,df_base2,how='inner',on='card_id')
df_base = p.merge(df_base,df_base3,how='inner',on='card_id')
df_base = p.merge(df_base,df_base4,how='inner',on='card_id')
```

```python
#CORRELATION MATRIX (Pearson Correlation to measure how similar are 2 solutions)
plt.figure(figsize=(16,12))
sns.heatmap(df_base.iloc[:,1:].corr(),annot=True,fmt=".2f")
```

```python
# ALTERNATIVE WAY - RMSE MATRIX (RMSE to measure how similar are 2 solutions)
M = np.zeros([df_base.iloc[:,1:].shape[1],df_base.iloc[:,1:].shape[1]])
for i in np.arange(M.shape[1]):
    for j in np.arange(M.shape[1]):
        M[i,j] = sqrt(metrics.mean_squared_error(df_base.iloc[:,i+1], df_base.iloc[:,j+1]))
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(M,annot=True,fmt=".2f")
```

```python
#SOLUTION = MEAN OF COLUMNS
df_base['target'] = df_base.iloc[:,1:].mean(axis=1)
df_base[['card_id','target']].to_csv("Bestoutput.csv",index=False)
```
